# Seaborn圖樣式
有五種預設seaborn主題：darkgrid，whitegrid，dark，white，和ticks。它們各自適合不同的應用和個人喜好。默認主題是darkgrid

```python
import numpy as np
import seaborn as sns
```

```python
#設定圖形樣式 - whitegrid
# use sns.set
sns.set_style("whitegrid")
# 利用 NUMPY 去建立資料集
# np.random.RandomState 設定數學式
rs=np.random.RandomState(10)

x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)


# 畫圖
# sns.residplot
sns.residplot(x, y);
```

繪製單變數分佈: 
快速檢視多個單變數分佈的最便捷方式是distplot() 函數。默認情況下，這將繪製直方圖並適合內核密度估計值(KDE)。

作業目標(2)\
(1)使用 distplot()使用簡單的規則來正確猜測預設情況下正確的數位，但嘗試更多或更少的 bin 可能會顯示資料中的其他特徵:\
(2)有無kde對圖形分布的影響

```python
# bin: 指的是特徵值, 
# kde: on/off
# sns.distplot();
sns.distplot(x, bins='auto');
# （1）ans: 使用auto自動生成10條bins，即可看出資料分布接近常態分配，如果用50條bins反而會使直方圖過於複雜、不易觀察。
```

```python
sns.distplot(x, bins=50);
```

```python
sns.distplot(x, bins=10);
# （2）ans: 有無KDE對圖形分佈並無影響，因為KDE只是在估計未知的密度函數，透過KDE可以比較直觀的看出數據樣本本身的分布特徵。
```

```python
sns.distplot(x, bins=10, kde=False);
```
